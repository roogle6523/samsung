# 삼성전자(우)주식에 관한 기사 크롤링


```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

url = "https://finance.naver.com/item/news_news.nhn?code=005935&page=1&sm=entity_id.basic&clusterId="
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")

soup
```

```python
#타이틀 접근
len(soup.find_all("a","tit"))

SS_title = [soup.find_all("a","tit")[n].get_text() for n in range(0,12)]

SS_title
```

```python
#언론사 데이터 접근
soup.find_all("td","info")
```

```python
#날짜 데이터 접근
soup.find_all("td","date")
```

```python
dataindex = np.arange(51)
dataindex = dataindex[1:]
dataindex
```

```python
SS_title = []
SS_press = []
SS_date = []
for i in tqdm_notebook(dataindex):
    url = "https://finance.naver.com/item/news_news.nhn?code=005935&page={pagenum}&sm=entity_id.basic&clusterId="
    page = urlopen(url.format(pagenum=(i)))
    soup = BeautifulSoup(page, "html.parser")
    
    cnt = len(soup.find_all("a","tit"))

    SS_title.extend([soup.find_all("a","tit")[n].get_text() for n in range(0,cnt)])
    SS_press.extend([soup.find_all("td","info")[n].get_text() for n in range(0,cnt)])
    SS_date.extend([soup.find_all("td","date")[n].get_text() for n in range(0,cnt)])
 
```

### 명사 빈도 카운트

```python
f = open('SS_title.txt', 'r',encoding = 'utf-8')
news = f.read()

noun = okt.nouns(news)

for i,v in enumerate(noun):
    if len(v)<2:
        noun.pop(i)
count = Counter(noun)


#명사 빈도 카운트
noun_list = count.most_common(100)
for v in noun_list:
    print(v)
```
