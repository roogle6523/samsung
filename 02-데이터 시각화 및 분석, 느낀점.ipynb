```python
f = open('SS_title.txt', 'r',encoding = 'utf-8')
news = f.read()


wordcloud=WordCloud(font_path="C:/Windows/Fonts/malgun.ttf",
                    background_color="white",
                    width = 1000,
                    height = 1000,
                    max_words = 200,
                    max_font_size= 200)
wordcloud.generate(news)
wordcloud.to_file('news_cloud2.png')

plt.figure(figsize=(12, 12))
plt.imshow(wordcloud)

plt.axis("off")# 범위 삭제
```
